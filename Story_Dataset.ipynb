```python
# Take a sample of 1000 rows
df_sample = df.sample(n=1000, random_state=42)

# Count number of stories per Genre in this sample
genre_counts = df_sample['Genre'].value_counts()

plt.figure(figsize=(4,4))

# Colors for each slice
colors = ['royalblue','orange', 'HotPink', 'forestgreen', 'red', 'mediumpurple']#lightcoral

# Create a pie chart
plt.pie(genre_counts, labels = genre_counts.index, colors = colors, autopct = '%1.1f%%', startangle=140)
plt.title('Genre Distribution')
plt.show()
```

**✅ Observation:-**
The genre distribution is fairly balanced in the sample, with Drama being the most represented and Mystery the least.

```python
# Take a sample of 1000 rows
df_sample = df.sample(n=1000, random_state=42)

# Take a sample of 1000 rows
df_sample = df.sample(n=1000, random_state=42)

# Group by Genre and sum Likes + Views
likes_per_genre = df_sample.groupby('Genre')['Likes'].sum()
views_per_genre = df_sample.groupby('Genre')['Views'].sum()

# Create subplots: 1 row, 2 columns
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # Wider figure for side-by-side

# Plot Likes
axs[0].bar(likes_per_genre.index, likes_per_genre.values, color='skyblue')
axs[0].set_xlabel("Genre")
axs[0].set_ylabel("Total Likes")
axs[0].set_title("Total Likes by Genre")

# Plot Views
axs[1].bar(views_per_genre.index, views_per_genre.values, color='lightcoral')
axs[1].set_xlabel("Genre")
axs[1].set_ylabel("Total Views")
axs[1].set_title("Total Views by Genre")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

**✅ Observation:-**
Drama leads in both total Likes and Views, indicating its strong popularity among readers in the sample.

```python
# Group by both Country and Genre, then sum Price
price_per_country_genre = df.groupby(['Country', 'Genre'])['Price'].sum()

# Convert to DataFrame
price_country_genre_df = price_per_country_genre.reset_index()

# Rename column
price_country_genre_df.columns = ['Country', 'Genre', 'Total Price']

# Sort by Total Price
price_country_genre_df = price_country_genre_df.sort_values(by='Total Price', ascending=False)

price_country_genre_df.head(10)
```

**✅ Observation:-**   UK Romance stories generated the highest total price, with Canada Drama and USA Horror closely following in sales value.

```python
# Group by Genre and sum Award
awards_per_genre = df.groupby('Genre')['Awards'].size()

# Convert to DataFrame
awards_genre_df = awards_per_genre.reset_index()

# Rename column
awards_genre_df.columns = ['Genre', 'Total Awards']

# Sort by Total Awards
awards_genre_df = awards_genre_df.sort_values(by='Total Awards', ascending=False)
awards_genre_df
```

**✅ Observation:-**
Horror genre has the highest total awards, showing it’s the most frequently awarded genre in the dataset.

```python
df_sample = df.sample(500)  # Creates a random sample of 500 rows
# Group by Format and sum Likes
likes_by_format = df_sample.groupby('Format')['Likes'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(6,4))
plt.bar(likes_by_format.index, likes_by_format.values, color='lightseagreen')
plt.xlabel('Format')
plt.ylabel('Total Likes')
plt.title('Total Likes by Format (Sample of 500)')
plt.show()
```

**✅ Observation:-**

```python
# Group by Format and sum Price
price_by_format = df_sample.groupby('Format')['Price'].sum().sort_values(ascending=True)

# Define a list of colors (or use colormap)
colors = plt.cm.Set3(range(len(price_by_format)))

#plot
plt.figure(figsize=(6,4))
plt.barh(price_by_format.index, price_by_format.values, color=colors)
plt.xlabel('Format')
plt.ylabel('Total Price')
plt.title('Total Price by Format (Sample of 500)')
plt.show()
```

**✅ Observation:-**  
Paperback format contributed the highest total price in the sample, indicating it generated more revenue compared to other formats.
